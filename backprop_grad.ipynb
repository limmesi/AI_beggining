```python
import torch
dimensions = [784,100,10]
N_batch = 64
```

```python
X = torch.randn(dimensions[0], N_batch)
y = torch.randn(dimensions[-1], N_batch)
```

```python
W1 = torch.randn(100, 784, requires_grad = True)
b1 = torch.ones(100, 1, requires_grad = True)
W2 = torch.randn(10, 100, requires_grad = True)
b2 = torch.ones(64, requires_grad = True)
```

```python
Z1 = torch.add(torch.mm(W1, X), b1)
A1 = torch.relu(Z1)
Z2 = torch.add(torch.mm(W2, A1), b2)
A2 = torch.relu(Z2)
```

```python
A2.shape
```

```python

loss = (torch.mean(A2, 1, True) - y)**2
loss = loss.mean()
```

```python
loss.backward()
```

```python
lr = 5e-6
```

```python
with torch.no_grad():
      W1 -= loss * lr
      b1 -= loss * lr
      W2 -= loss * lr
      b2 -= loss * lr
      
      W1.grad.zero_()
      W2.grad.zero_()
      b1.grad.zero_()
      b2.grad.zero_()
```

```python
epochs = 600
```

```python
W1 = torch.randn(100, 784, requires_grad = True)
b1 = torch.ones(100, 1, requires_grad = True)
W2 = torch.randn(10, 100, requires_grad = True)
b2 = torch.ones(64, requires_grad = True)
```

```python
for epoch in range(epochs):
    Z1 = torch.add(torch.mm(W1, X), b1)
    A1 = torch.relu(Z1)
    Z2 = torch.add(torch.mm(W2, A1), b2)
    A2 = torch.relu(Z2)

    loss = (torch.mean(A2, 1, True) - y)**2
    loss = loss.mean()  
    loss.backward()

    if epoch % 50 == 0:
      print(loss)

    with torch.no_grad():
        W1 -= loss * lr
        b1 -= loss * lr
        W2 -= loss * lr
        b2 -= loss * lr
        W1.grad.zero_()
        W2.grad.zero_()
        b1.grad.zero_()
        b2.grad.zero_()
```
